## Reference Variable:

When we create object, on that time the variable in which we are storing object is known as Reference variable.

## Pass By Reference

```python
class Customer:

  def __init__(self,name,gender):
    self.name = name
    self.gender = gender

def greet(customer):
  if customer.gender == "Male":
    print("Hello", customer.name, "sir")

  else:
    print("Hello", customer.name, "ma'am")
```

```python
cust = Customer("Ankita", "Female")
```

```python
new_cust = greet(cust)
```

- We can pass a object of class in function argument same as other data or arguments.

```python
class Customer:

  def __init__(self,name):
    self.name = name

def greet(customer):
  print(id(customer))
```

```python
cust = Customer("Ankita")
print(id(cust))
```

```python
greet(cust)
```

- I created a Customer class
- then i created an object for Customer class
- and then i passed that object (Reference Variable) in a greet function
- and you can see that both ids are same, it means they both are pointing to same object. 

- It is working like aliasing

```python
class Customer:

  def __init__(self,name):
    self.name = name

def greet(customer):
  customer.name = "Upendra"
  print(customer.name)
```

```python
cust = Customer("Ankita")

greet(cust)

print(cust.name)
```

- from above we can say that if we pass an object to any function and made any chages to that object inside function then our original object will also change.

```python
class Customer:

  def __init__(self,name):
    self.name = name

def greet(customer):
  print(id(customer))
  customer.name = "Upendra"
  print(customer.name)
  print(id(customer))
```

```python
cust = Customer("Ankita")
print(id(cust))

greet(cust)

print(cust.name)
```

- Here we are able to change object with the help of function and also they are in same memory location, means before and after change object id is same. It means class objects are mutable like lists, dict and sets.

- If you are able to edit any data types or object in same address(id), then that data type or object is mutable.

```python
def change(L):
  print(id(L))
  L.append(5)
  print(id(L))
```

```python
L1 = [1,2,3,4]
print(id(L1))
print(L1)
```

```python
change(L1)
```

```python
print(L1)
```

- List is mutable, here we are able to make changes in list in same address(id).

- It is risky some timee to send list or object directly to any function, here to avoid this we can do clonning

```python
L2 = [5,6,7,8]
print(L2)
print(id(L2))
```

```python
change(L2[:])
```

```python
print(L2)
```

- So what we learn from here that if we are taking function call and putting class object as function argument then we should use clonnig, so that function will not able to make any permannets changes.



- If do the same thing in a tuple then this will not happen because tuple is immutable.

```python
T1 = (1,2,3,4)
```

```python
print(T1)
print(id(T1))
```

```python
def change(T):
  print(id(T))
  T = T + (5,6)
  print(id(T))
```

```python
change(T1)
```

```python
print(T1)
```

- So the concept is:

 - If we are performing pass by reference and giving mutable data type then our original object will get changed. It means permanent change will happen.

 - If we are giving any immutable data type then our original object will not get changed. permanent change will not happen.

## Collection of objects

```python
class Customer:

  def __init__(self,name, age):
    self.name = name
    self.age = age
  
  def intro(self):
    print("I am", self.name, "and I am", self.age, "years old")
```

```python
c1 = Customer("Nitish", 34)
c2 = Customer("Ankit", 45)
c3 = Customer("Neha", 32)
```

```python
L = [c1, c2, c3]
```

```python
for i in L:
  print(i)
```

```python
for i in L:
  print(i.name, i.age)
```

```python
for i in L:
  i.intro()
```
